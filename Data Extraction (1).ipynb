```python
import json
import os
import pandas as pd
po = []
ne = []
li = []
for i in range(7,19):
    if(i<10):
         source = 'C:/Users/CU696Aliasgar/Downloads/Polraris-PolarisEtech-PolarisStereo_11-20 Aug JSONs/Polraris-PolarisEtech-PolarisStereo/2022/08/15/0'+str(i)
    else:
        source = 'C:/Users/CU696Aliasgar/Downloads/Polraris-PolarisEtech-PolarisStereo_11-20 Aug JSONs/Polraris-PolarisEtech-PolarisStereo/2022/08/15/'+str(i)
    q = os.listdir(source)
    print(len(q))

    for fi in q:
        al = source +'/' + fi
        with open(al) as f:
            try:
                data = json.load(f)
                li.append((data.get("app_data")).get("url"))
                po.append(data.get("sentiment"))
                ne.append(fi[0:10])
                if(len(po)!=len(ne)):
                    print(fi)
            
            except:
                print(fi)
```
